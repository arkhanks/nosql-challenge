### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = establishments.find({'scores.Hygiene': 20})
query_result_list = list(query)  # Convert cursor to list

# Use count_documents to display the number of documents in the result
hygieneCount = establishments.count_documents({'scores.Hygiene': 20})
print(f'Number of establishments with hygiene score of 20: {hygieneCount}')

# Display the first document in the results using pprint
first_document = query_result_list[0] if query_result_list else None
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame by converting the cursor to a list first

query_df = pd.DataFrame(query_result_list)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(query_df))

# Display the first 10 rows of the DataFrame
query_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Fetch a sample document to inspect the field names and data structure
sample_document = establishments.find_one()  # Fetches just one document
pprint(sample_document)
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_london = establishments.find({
    'LocalAuthorityName': {'$regex': 'London', '$options': 'i'},
    'RatingValue': {'$gte': 4}
})
# Put query results in a list
london_result_list = list(query_london)

# Use count_documents to display the number of documents in the result
London4Count = establishments.count_documents({
    'LocalAuthorityName': {'$regex': 'London', '$options': 'i'},
    'RatingValue': {'$gte': 4}
})
print(f'Number of establishments with LocalAuthorityName London and RatingValue >= 4: {London4Count}')

# Display the first document in the results using pprint
first_document = london_result_list[0] if london_result_list else None
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(london_result_list)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(london_df))

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Check if there are establishments with RatingValue = 5
query_rating_value_5 = establishments.find({'RatingValue': 5})
rating_value_5_count = establishments.count_documents({'RatingValue': 5})

print(f"Number of establishments with RatingValue = 5: {rating_value_5_count}")
```

```python
# Check if geocode data exists and is in the correct format
query_geocode_exists = establishments.find({'geocode': {'$exists': True}})
geocode_count = establishments.count_documents({'geocode': {'$exists': True}})
print(f"Number of establishments with geocode data: {geocode_count}")
```

```python

query_rating_value = establishments.find({
    'geocode': {
        '$geoWithin': {
            '$centerSphere': [
                [longitude, latitude],  # Correct coordinate order [longitude, latitude]
                5.0 / 6378.1  # The radius in radians (dividing by Earth's radius in kilometers)
            ]
        }
    },
    'RatingValue': {'$exists': True}
}).limit(5)

rating_value_count = establishments.count_documents({
    'geocode': {
        '$geoWithin': {
            '$centerSphere': [
                [longitude, latitude],  # Correct coordinate order [longitude, latitude]
                5.0 / 6378.1  # The radius in radians (dividing by Earth's radius in kilometers)
            ]
        }
    },
    'RatingValue': {'$exists': True}
})

print(f"Establishments near 'Penang Flavours' with RatingValue: {rating_value_count}")

```

```python
query = establishments.find({
    'geocode': {
        '$geoWithin': {
            '$centerSphere': [
                [longitude, latitude],
                5.0 / 6378.1
            ]
        }
    },
    'RatingValue': 5,  # Filtering for RatingValue of 5
    'scores.Hygiene': {'$exists': True, '$ne': '', '$gt': 0}  # Ensure hygiene score is non-zero
}).sort('scores.Hygiene', 1).limit(5)  # Sort by lowest Hygiene score

query_result_list = list(query)

# Print the results
print("Top 5 establishments with RatingValue 5, sorted by lowest non-zero Hygiene score:")
for doc in query_result_list:
    pprint(doc)
```

```python
# Handling Missing or Zero Hygiene Scores: If you are looking for establishments that have a hygiene score 
# recorded (whether it's positive or not), you could instead include a check for missing values (e.g., None or empty string '') 
# and filter out records that don't have actual scores.
query = establishments.find({
    'geocode': {
        '$geoWithin': {
            '$centerSphere': [
                [longitude, latitude],
                5.0 / 6378.1
            ]
        }
    },
    'RatingValue': 5,  # Filtering for RatingValue of 5
    'scores.Hygiene': {'$exists': True, '$ne': ''}  # Ensure hygiene score exists and is not empty
}).sort('scores.Hygiene', 1).limit(5)  # Sort by lowest Hygiene score

query_result_list = list(query)

# Print the results
print("Top 5 establishments with RatingValue 5, sorted by lowest Hygiene score:")
for doc in query_result_list:
    pprint(doc)
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01

# Coordinates for the "Penang Flavours" restaurant
latitude = 51.49014200
longitude = 0.08384000

# Create a 2dsphere index on 'geocode' (if not already created)
establishments.create_index([("geocode", "2dsphere")])      

query = establishments.find({
    'geocode': {
        '$nearSphere': {
            '$geometry': {
                'type': 'Point',
                'coordinates': [longitude, latitude]  
            },
            '$maxDistance': degree_search  
        }
    },
    'RatingValue': 5  # Filter establishments with RatingValue 5
}).sort('scores.Hygiene', 1).limit(5)  # Sort by Hygiene score (ascending) and limit to 5 results

query_result_list = list(query)

# Print the results
print("Top 5 establishments with RatingValue 5, sorted by lowest Hygiene score:")
for doc in query_result_list:
    pprint(doc)
```
