### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = establishments.find({'scores.Hygiene': 20})
query_result_list = list(query)  # Convert cursor to list

# Use count_documents to display the number of documents in the result
hygieneCount = establishments.count_documents({'scores.Hygiene': 20})
print(f'Number of establishments with hygiene score of 20: {hygieneCount}')

# Display the first document in the results using pprint
first_document = query_result_list[0] if query_result_list else None
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame by converting the cursor to a list first

query_df = pd.DataFrame(query_result_list)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(query_df))

# Display the first 10 rows of the DataFrame
query_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query_london = establishments.find({
    'LocalAuthorityName': {'$regex': 'London', '$options': 'i'},
    'RatingValue': {'$gte': 4}
})
# Put query results in a list
london_result_list = list(query_london)

# Use count_documents to display the number of documents in the result
London4Count = establishments.count_documents({
    'LocalAuthorityName': {'$regex': 'London', '$options': 'i'},
    'RatingValue': {'$gte': 4}
})
print(f'Number of establishments with LocalAuthorityName London and RatingValue >= 4: {London4Count}')

# Display the first document in the results using pprint
first_document = london_result_list[0] if london_result_list else None
pprint(first_document)
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(london_result_list)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(london_df))

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
degree_search = 0.01

# Coordinates for the "Penang Flavours" restaurant
latitude = 51.49014200
longitude = 0.08384000

query = establishments.find({
    '$and': [
        {'geocode.latitude': {'$gte': latitude -0.01, '$lte': latitude + 0.01}},
        {'geocode.longitude': {'$gte': longitude -0.01, '$lte': longitude + 0.01}},
        {'RatingValue': 5},  # Filtering for RatingValue of 5
        {'scores.Hygiene': {'$gt': 0, '$exists': True, '$ne': ''}}  # Ensure hygiene score exists and is not empty
        ]
}).sort('scores.Hygiene', 1).limit(10)  # Sort by lowest Hygiene score

query_result_list = list(query) # convert to list for easier processing

# Print the results
print("Top 5 establishments with RatingValue 5, sorted by lowest Hygiene score:")
for doc in query_result_list:
    pprint(doc)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:

# Uses $match to filter for establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene':0}}

# Groups the matches by Local Authority and counts the number of establishments in each Local Authority
group_query = {'$group': {'_id': '$LocalAuthorityName', 'count':{'$sum': 1}}}

# Sorts the matches from highest to lowest
sort_values = {'$sort': {'count': -1}}

# Limit to the top 10 Local Authorities
limit = {'$limit': 10}

# Put the pipeline together
pipeline = [match_query, group_query, sort_values, limit]

# Run the pipeline throught the aggregate method and save results to a variable
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Top 10 Local Authority areas with establishments having hygiene score of 0:")

# Print the first 10 results
for doc in results:
    print(f"Local Authority: {doc['_id']}, Count: {doc['count']}")
```

```python
# Convert the result to a Pandas DataFrame
local_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(local_df))

# Display the first 10 rows of the DataFrame
local_df.head(10)
```
